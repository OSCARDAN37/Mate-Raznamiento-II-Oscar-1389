### $Suma/Resta$
* Mismo signo: Se suman los numeros y se conserva el signo
* Signo diferente : Se restan los numeros y se conserva el signo que sea dominante
### $Multiplicacion/Division$

* signos iguales: El resultado es posiivo
* signos diferentes: El resultado es negativo

# $Problema$ $1$:

```python
# Problema 1:
# x + 8 = 15
suma = int(input("Ingresa la cantidad que se suma: "))
total = int(input("ingresa el resultado de la operacion: "))

X = total - suma
print(f"El valor de X es: {X}")
```

# $Problema$ $3$:

```python
# Problema 3:
# 3x + 4x = 21
x1 = int(input("Ingresa la canridad por la que se multiplica la primer variable: "))
x2 = int(input("Ingresa la canridad por la que se multiplica la segunda variable: "))
R = int(input("Ingresa el resultado de la operacion: "))

X = R / (x1 + x2)
print(f"El valor de X es: {X}")
```

# $Problema$ $5$:

```python
# Problema 5:
# 6a - 2a + a = 20
a1 = int(input("Ingresa el primer coeficiente de a (6): "))
a2 = int(input("Ingresa el segundo coeficiente de a (2): "))
a3 = int(input("Ingresa el tercer coeficiente de a (1): "))
R = int(input("Ingresa el resultado de la operación (20): "))

# Simplificamos: (6 - 2 + 1) = 5
A = R / (a1 - a2 + a3)
print(f"El valor de a es: {A}")
```

# $Problema$ $7$:

```python
# Problema 7:
# 4x + 3 = 2x + 11

# Datos del lado izquierdo (4x + 3)
coeficiente_izq = int(input("Ingresa el coeficiente de x de la izquierda (4): "))
numero_izq = int(input("Ingresa el número que suma/resta a la izquierda (3): "))

# Datos del lado derecho (2x + 11)
coeficiente_der = int(input("Ingresa el coeficiente de x de la derecha (2): "))
numero_der = int(input("Ingresa el número que suma/resta a la derecha (11): "))

# Aplicando el despeje: x = (D - B) / (A - C)
# Primero restamos los números y luego dividimos por la resta de los coeficientes
X = (numero_der - numero_izq) / (coeficiente_izq - coeficiente_der)

print(f"\nEl valor de X es: {X}")
```

# $Problema$ $9$:

```python
# Problema 9:
# 9m + 4 = 5m + 20

# Lado izquierdo (9m + 4)
m1 = int(input("Ingresa el coeficiente de m de la izquierda (9): "))
n1 = int(input("Ingresa el número que acompaña a la izquierda (4): "))

# Lado derecho (5m + 20)
m2 = int(input("Ingresa el coeficiente de m de la derecha (5): "))
n2 = int(input("Ingresa el número que acompaña a la derecha (20): "))

M = (n2 - n1) / (m1 - m2)

print(f"\nEl valor de m es: {M}")
```

# $Problema$ $11$:

```python
# Problema 11:
# 6a + 5 = 2a + 21

# Datos del lado izquierdo (6a + 5)
coef_a_izq = int(input("Ingresa el coeficiente de a de la izquierda (6): "))
num_izq = int(input("Ingresa el número de la izquierda (5): "))

# Datos del lado derecho (2a + 21)
coef_a_der = int(input("Ingresa el coeficiente de a de la derecha (2): "))
num_der = int(input("Ingresa el número de la derecha (21): "))

# Proceso de despeje: (21 - 5) / (6 - 2)
a = (num_der - num_izq) / (coef_a_izq - coef_a_der)

print(f"\nEl valor de a es: {a}")
```

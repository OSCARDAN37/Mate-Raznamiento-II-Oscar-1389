### $Suma/Resta$
* Mismo signo: Se suman los numeros y se conserva el signo
* Signo diferente : Se restan los numeros y se conserva el signo que sea dominante
### $Multiplicacion/Division$

* signos diferentes: El resultado es negativo

# $Problema$ $1$:

1.  **Planteamiento:** Tenemos una suma donde desconocemos uno de los valores.
    $$x + 8 = 15$$
2.  **Despeje de la variable:** Para mover el $+8$ al otro lado del signo igual ($=$), debemos aplicar la operación contraria, que es la resta.
    $$x = 15 - 8$$
3.  **Solución:** Al realizar la operación, obtenemos el valor de la incógnita.
    $$x = 7$$


```python
# Problema 1:
# x + 8 = 15
suma = int(input("Ingresa la cantidad que se suma: "))
total = int(input("ingresa el resultado de la operacion: "))

X = total - suma
print(f"El valor de X es: {X}")
```

# $Problema$ $3$:

1.  **Planteamiento:** Tenemos dos términos que contienen la misma incógnita.
    $$3x + 4x = 21$$
2.  **Simplificación:** Sumamos los coeficientes (los números que acompañan a la $x$). Como $3 + 4 = 7$, la ecuación se reduce a:
    $$7x = 21$$
3.  **Despeje de la variable:** El número 7 está multiplicando a la $x$. Para despejarla, pasamos el 7 al otro lado dividiendo:
    $$x = \frac{21}{7}$$
4.  **Solución:** $$x = 3$$


```python
# Problema 3:
# 3x + 4x = 21
x1 = int(input("Ingresa la canridad por la que se multiplica la primer variable: "))
x2 = int(input("Ingresa la canridad por la que se multiplica la segunda variable: "))
R = int(input("Ingresa el resultado de la operacion: "))

X = R / (x1 + x2)
print(f"El valor de X es: {X}")
```

# $Problema$ $5$:
### Problema: $6a - 2a + a = 20$
1.  **Simplificación:** Sumamos y restamos las letras: $6 - 2 + 1 = 5$.
    * Ecuación simplificada: $5a = 20$
2.  **Despeje:** $a = 20 / 5$
3.  **Resultado:** $a = 4$

```python
# Problema 5:
# 6a - 2a + a = 20
a1 = int(input("Ingresa el primer coeficiente de a : "))
a2 = int(input("Ingresa el segundo coeficiente de a : "))
a3 = int(input("Ingresa el tercer coeficiente de a : "))
R = int(input("Ingresa el resultado de la operación : "))

# Simplificamos: (6 - 2 + 1) = 5
A = R / (a1 - a2 + a3)
print(f"El valor de a es: {A}")
```

# $Problema$ $7$:
### Problema: $4x + 3 = 2x + 11$
1.  **Agrupación:** Pasamos las $x$ a la izquierda y los números a la derecha:
    * $4x - 2x = 11 - 3$
2.  **Simplificación:** $2x = 8$
3.  **Despeje:** $x = 8 / 2$
4.  **Resultado:** $x = 4$





```python
# Problema 7:
# 4x + 3 = 2x + 11

# Datos del lado izquierdo (4x + 3)
coeficiente_izq = int(input("Ingresa el coeficiente de x de la izquierda : "))
numero_izq = int(input("Ingresa el número que suma/resta a la izquierda : "))

# Datos del lado derecho (2x + 11)
coeficiente_der = int(input("Ingresa el coeficiente de x de la derecha : "))
numero_der = int(input("Ingresa el número que suma/resta a la derecha : "))

# Aplicando el despeje: x = (D - B) / (A - C)
# Primero restamos los números y luego dividimos por la resta de los coeficientes
X = (numero_der - numero_izq) / (coeficiente_izq - coeficiente_der)

print(f"\nEl valor de X es: {X}")
```

# $Problema$ $9$:
### Problema: $9m + 4 = 5m + 20$
1.  **Agrupación:** $9m - 5m = 20 - 4$
2.  **Simplificación:** $4m = 16$
3.  **Despeje:** $m = 16 / 4$
4.  **Resultado:** $m = 4$

```python
# Problema 9:
# 9m + 4 = 5m + 20

# Lado izquierdo (9m + 4)
m1 = int(input("Ingresa el coeficiente de m de la izquierda : "))
n1 = int(input("Ingresa el número que acompaña a la izquierda : "))

# Lado derecho (5m + 20)
m2 = int(input("Ingresa el coeficiente de m de la derecha : "))
n2 = int(input("Ingresa el número que acompaña a la derecha : "))

M = (n2 - n1) / (m1 - m2)

print(f"\nEl valor de m es: {M}")
```

# $Problema$ $11$:
### Problema: $6a + 5 = 2a + 21$
1.  **Agrupación:** $6a - 2a = 21 - 5$
2.  **Simplificación:** $4a = 16$
3.  **Despeje:** $a = 16 / 4$
4.  **Resultado:** $a = 4$


```python
# Problema 11:
# 6a + 5 = 2a + 21

# Datos del lado izquierdo (6a + 5)
coef_a_izq = int(input("Ingresa el coeficiente de a de la izquierda : "))
num_izq = int(input("Ingresa el número de la izquierda : "))

# Datos del lado derecho (2a + 21)
coef_a_der = int(input("Ingresa el coeficiente de a de la derecha : "))
num_der = int(input("Ingresa el número de la derecha : "))

# Proceso de despeje: (21 - 5) / (6 - 2)
a = (num_der - num_izq) / (coef_a_izq - coef_a_der)

print(f"\nEl valor de a es: {a}")
```

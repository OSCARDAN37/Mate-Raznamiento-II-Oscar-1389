### Problema 1: La suma de un número más 9 es igual a 21
Para resolver esto, sigamos un orden lógico:

1.  **Planteamiento:** Identificamos la incógnita. Llamaremos $x$ al "número desconocido".
2.  **Traducción algebraica:** "La suma de un número ($x$) más 9 ($+ 9$) es igual ($=$) a 21".
    * Ecuación: $x + 9 = 21$
3.  **Despeje de la variable:** Para dejar sola a la $x$, restamos 9 en ambos lados de la igualdad:
    * $x = 21 - 9$
4.  **Solución:** * $x = 12$


```python
# Problema 1: Encontrar el numero desconocido

resultado_total = int(input("Ingrese el resultado total: "))
numero_conocido = int(input("Ingrese el número conocido: "))

incognita_x = resultado_total - numero_conocido

print(f"El número buscado en el Problema 1 es: {incognita_x}")
```


### Problema 2: Las edades de Pedro y Ana
Este problema requiere relacionar dos variables basándonos en una sola.

1.  **Planteamiento:**
    * Edad de Ana: $x$
    * Edad de Pedro: $2x$ (porque es el doble que la de Ana).
2.  **Traducción algebraica:** "Entre ambos suman un total de 36 años".
    * Ecuación: $x + 2x = 36$
3.  **Simplificación:** Sumamos las términos semejantes:
    * $3x = 36$
4.  **Despeje de la variable:** El 3 que multiplica a la $x$ pasa dividiendo:
    * $x = \frac{36}{3}$
    * $x = 12$
5.  **Solución final:**
    * Edad de **Ana** ($x$): **12 años**.
    * Edad de **Pedro** ($2x$): $2(12) = \mathbf{24 \text{ años}}$.


```python
# Problema 2: x + 2x = 36  => 3x = 36

suma_total = int(input("Ingrese la suma total de las edades: "))


edad_ana = suma_total / 3

edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```

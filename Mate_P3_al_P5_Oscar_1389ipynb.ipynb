### Problema 3: Un número menos 7 es igual a 18
1.  **Planteamiento:** Sea $x$ el número buscado.
2.  **Traducción:** "Un número ($x$) al que se le resta siete ($- 7$) da como resultado ($=$) 18".
    * Ecuación: $x - 7 = 18$
3.  **Despeje:** El 7 está restando, pasa al otro lado sumando:
    * $x = 18 + 7$
4.  **Solución:** $x = 25$


```python
resta = int(input("Ingrese el número que resta: "))
resultado = int(input(f"Ingrese el resultado de x - {resta}: "))

# Para hallar x, sumamos lo que se restó
numero = resultado + resta

print(f" X es igual a: {numero}")
```

### Problema 4: El triple de un número más 4 es igual a 31
1.  **Planteamiento:** Sea $x$ el número. El "triple" se representa como $3x$.
2.  **Traducción:** "El triple de un número ($3x$) más 4 ($+ 4$) es igual ($=$) a 31".
    * Ecuación: $3x + 4 = 31$
3.  **Despeje:**
    * Primero quitamos el 4: $3x = 31 - 4$ $\rightarrow$ $3x = 27$
    * Luego quitamos el 3 (pasa dividiendo): $x = \frac{27}{3}$
4.  **Solución:** $x = 9$

```python
# Problema 4: 3x + 4 = 31
total = int(input("Ingrese el total: "))
termino_independiente = int(input("Ingrese el término independiente: "))
coeficiente = int(input("Ingrese el coeficiente: "))

paso1 = total - termino_independiente

x = paso1 / coeficiente

print(f"X es igual a: {int(x)}")
```

### Problema 5: Dimensiones de un terreno rectangular
1.  **Planteamiento:**
    * Ancho: $x$
    * Largo: $x + 5$ (cinco más que el ancho).
2.  **Traducción:** El perímetro de un rectángulo es la suma de sus cuatro lados: $P = 2(\text{largo}) + 2(\text{ancho})$.
    * Ecuación: $2(x + 5) + 2(x) = 50$
3.  **Despeje:**
    * Multiplicamos: $2x + 10 + 2x = 50$
    * Sumamos las $x$: $4x + 10 = 50$
    * Restamos 10: $4x = 40$
    * Dividimos entre 4: $x = \frac{40}{4} = 10$
4.  **Solución:**
    * **Ancho:** 10 m.
    * **Largo:** $10 + 5 = \mathbf{15 \text{ m}}$.


```python
# Pedir el perímetro al usuario
p = float(input("¿Cuánto mide el perímetro total?: "))

# P = 2(x + 5) + 2x  => P = 4x + 10  => 4x = P - 10 => x = (P - 10) / 4
ancho = (p - 10) / 4
largo = ancho + 5

# Mostrar resultados
print(f"Cada ancho mide: {ancho} m")
print(f"Cada largo mide: {largo} m")
```
